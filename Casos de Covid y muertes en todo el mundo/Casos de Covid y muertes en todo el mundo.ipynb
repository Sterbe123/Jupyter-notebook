# Casos de Covid y muertes en todo el mundo

## Acerca del conjunto de datos

La enfermedad por coronavirus (COVID-19) es una enfermedad infecciosa causada por el virus SARS-CoV-2 .

La mayoría de las personas infectadas con el virus experimentarán una enfermedad respiratoria de leve a moderada y se recuperarán sin requerir un tratamiento especial. Sin embargo, algunos se enfermarán gravemente y requerirán atención médica. Las personas mayores y aquellas con afecciones médicas subyacentes, como enfermedades cardiovasculares, diabetes, enfermedades respiratorias crónicas o cáncer, tienen más probabilidades de desarrollar enfermedades graves. Cualquiera puede enfermarse de COVID-19 y enfermarse gravemente o morir a cualquier edad.

La mejor manera de prevenir y ralentizar la transmisión es estar bien informado sobre la enfermedad y cómo se propaga el virus. Protéjase a sí mismo y a los demás de las infecciones manteniéndose al menos a 1 metro de distancia de los demás, usando una mascarilla que le quede bien ajustada y lavándose las manos o usando un desinfectante a base de alcohol con frecuencia. Vacúnese cuando sea su turno y siga las indicaciones locales.

El virus puede propagarse desde la boca o la nariz de una persona infectada en pequeñas partículas líquidas cuando tose, estornuda, habla, canta o respira. Estas partículas van desde gotas respiratorias más grandes hasta aerosoles más pequeños. Es importante practicar la etiqueta respiratoria, por ejemplo, toser con el codo flexionado, y quedarse en casa y aislarse hasta que se recupere si no se siente bien.

¿Dónde siguen siendo altos los casos?

Los casos globales diarios cayeron después de un pico en la primavera, pero ahora están aumentando nuevamente, con la aparición de las subvariantes BA.4 y BA.5 de la variante Omicron .

Los estudios sugieren que Omicron, que rápidamente se volvió dominante en numerosos países, es más leve que la variante Delta, pero mucho más contagiosa. Las subvariantes son aún más contagiosas.

### 1.2 Verificamos el promedio de test por persona

```python
# eliminamos los datos nan
data_frame_paises_promedio = data_frame.copy()
data_frame_paises_promedio.dropna(subset=["Total Test", "Population", "Total Deaths", "Total Recovered"], inplace=True)
data_frame_paises_promedio
```

```python
#funcion para formater los datos
def formatear_datos(numero):
    return '{:,}'.format(numero).replace(",", ".")
```

```python
#obtenemos los promedio por paises
data_promedio_test_persona = []
for index in range(255):
    if index in data_frame_paises_promedio.index:
        data_promedio_test_persona.append([
        data_frame_paises_promedio["Country"][index], 
        (int(data_frame_paises_promedio["Total Test"][index].replace(",", ""))
        /
        int(data_frame_paises_promedio["Population"][index].replace(",", ""))),
        (int(data_frame_paises_promedio["Total Deaths"][index].replace(",", "")) * 100) 
        / 
        int(data_frame_paises_promedio["Total Cases"][index].replace(",", "")),
        (int(data_frame_paises_promedio["Total Deaths"][index].replace(",", "")) * 100) 
        / 
        int(data_frame_paises_promedio["Population"][index].replace(",", "")),
        (int(data_frame_paises_promedio["Total Recovered"][index].replace(",", "")) * 100) 
        / 
        int(data_frame_paises_promedio["Total Cases"][index].replace(",", ""))
        ])

# creamos las etiquetas para el nuevo data frame
columnas = ["Países", "Promedio de test por persona", "Mortalidad por casos activos",
            "Mortalidad por total de la población", "Recuperación"]

#creamos el nuevo data frame
data_frame_promedio = pd.DataFrame(data_promedio_test_persona, columns = columnas)

#redondeamos los promedio de test por persona
data_frame_promedio = data_frame_promedio.round({"Promedio de test por persona" : 2,
                                                 "Mortalidad por casos activos" : 2,
                                                 "Mortalidad por total de la población" : 2,
                                                 "Recuperación" : 2})

#datos
total_casos = data_frame_paises_promedio["Total Cases"].replace(",","", regex = True).astype(int).sum()
total_muertes = data_frame_paises_promedio["Total Deaths"].replace(",","", regex = True).astype(int).sum()
total_test = data_frame_paises_promedio["Total Test"].replace(",","", regex = True).astype(int).sum()
casos_activos = int(data_frame_paises_promedio["Active Cases"].replace(",","", regex = True).astype(float).sum(skipna = True))
pacientes_recuperado = int(data_frame_paises_promedio["Total Recovered"].replace(",","", regex = True).astype(float).sum(skipna = True))
print("Estos datos son apartir de la fecha de recopilación de este DataFrame")
print("Total de casos: ", formatear_datos(total_casos))
print("Casos Activos:", formatear_datos(casos_activos))
print("Total de test realizados en todo el mundo:", formatear_datos(total_test))
print("Pacientes Recuperados hasta la fecha:", formatear_datos(pacientes_recuperado))
print("Muertes en total:", formatear_datos(total_muertes))
```
