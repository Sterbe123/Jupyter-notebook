### Descripción
Shop Customer Data es un análisis detallado de los clientes ideales de una tienda imaginativa. Ayuda a una empresa a comprender mejor a sus clientes. El propietario de una tienda obtiene información sobre los Clientes a través de tarjetas de membresía.

El conjunto de datos consta de 2000 registros y 8 columnas:

* Identificación del cliente
* Género
* Edad
* Ingresos anuales
* Puntaje de gasto: puntaje asignado por la tienda, basado en el comportamiento del cliente y la naturaleza del gasto
* Profesión
* Experiencia Laboral - en años
* Tamaño de la familia

### 1.2 Análisis con gráficos de líneas

* Representación del gasto promedio según la edad 
* Vamos a procesar nuevos datos

```python
# creamos un nuevo data frame
data_frame_new = data_frame[["Gender","Age","Annual Income ($)"]]

# procesando los datos y creamos un nuevo data frame
data = []
for index in range((data_frame['Age'].max()+1)):
    suma_hombre = 0
    suma_mujeres = 0
    conteo_hombre = 0
    conteo_mujeres = 0
    for i in range(len(data_frame_new)):
        if data_frame_new["Age"][i] == index:
            if data_frame_new["Gender"][i] == "Male":
                suma_hombre = suma_hombre + data_frame_new["Annual Income ($)"][i]
                conteo_hombre = conteo_hombre + 1
            elif data_frame_new["Gender"][i] == "Female":
                suma_mujeres = suma_mujeres + data_frame_new["Annual Income ($)"][i]
                conteo_mujeres = conteo_mujeres + 1
                
    if conteo_mujeres != 0 and conteo_hombre != 0:
        data.append([index, round((suma_mujeres/conteo_mujeres),2), 
                     round((suma_hombre/conteo_hombre),2), 
                     suma_mujeres , 
                     suma_hombre,
                     conteo_mujeres,
                     conteo_hombre])
    else:
        data.append([index, 0, 0])
    
columnas = ["Años", "Gastos Promedio Mujeres", "Gastos Promedio Hombres", 
            "Gasto total mujeres", "Gasto total hombres",
            "Cantidad Mujeres",
            "Cantidad Hombres"]  
data_frame_filtrado = pd.DataFrame(data, columns=columnas) 
print("Cantidad de mujeres: " ,data_frame_filtrado["Cantidad Mujeres"].sum())
print("Gasto total mujeres: $" ,suma_mujeres)
print("Cantidad de hombres: " ,data_frame_filtrado["Cantidad Hombres"].sum())
print("Gasto total hombres: $" ,suma_hombre)
data_frame_filtrado.head()
```

```python
#comprobamos el monto registrado
data_frame['Annual Income ($)'].sum() == (data_frame_filtrado['Gasto total hombres'].sum() + 
                                          data_frame_filtrado['Gasto total mujeres'].sum())
```

```python
#instanciar gráfico de lineas
fig,ax = plt.subplots()
ax.plot(data_frame_filtrado["Años"].iloc[:50], data_frame_filtrado["Gastos Promedio Mujeres"].iloc[:50],
        marker="o", color="tab:purple")
ax.plot(data_frame_filtrado["Años"].iloc[:50], data_frame_filtrado["Gastos Promedio Hombres"].iloc[:50],
        marker="o" ,color="tab:blue")
ax.set_title("Representación de los gastos promedio hasta los 49 años.", loc = "center", fontsize=15)
plt.ylabel("Gastos", fontsize = 15)
plt.xlabel("Años", fontsize = 15)
ax.legend(labels = ["Mujeres","Hombres"],bbox_to_anchor = (1.1,1.0), fontsize=20)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(data_frame_filtrado["Años"].iloc[50:], data_frame_filtrado["Gastos Promedio Mujeres"].iloc[50:],
        marker="o", color="tab:purple")
ax.plot(data_frame_filtrado["Años"].iloc[50:], data_frame_filtrado["Gastos Promedio Hombres"].iloc[50:],
        marker="o" ,color="tab:blue")
ax.set_title("Representación de los gastos promedio desde los 49 años hasta los 99 años.", loc = "center", fontsize=15)
plt.ylabel("Gastos", fontsize = 15)
plt.xlabel("Años", fontsize = 15)
ax.legend(labels = ["Mujeres","Hombres"],bbox_to_anchor = (1.1,1.0), fontsize=20)
plt.show()
```

## 2. Transformación de datos (Gender y Profession)

```python
#importamos algunas librerias
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
import seaborn as sns

#copiamos el data frame original
data_frame_copy = data_frame.copy()

#instanciar el objeto
label_encoder = LabelEncoder()
imputer = SimpleImputer(strategy="median")

# convertimos los valores del Gender (género), Hombres se le asigna el valor de 1 y a las mujeres el valor de 0
data_frame_copy['Gender'] = label_encoder.fit_transform(data_frame_copy['Gender'])
data_frame_copy['Profession'] = label_encoder.fit_transform(data_frame_copy['Profession'])

#copiamos los valores numerico de Profession
data_frame_copy_num = data_frame_copy.select_dtypes(exclude='object')

# se le proporciona los valores numericos al imputer
imputer.fit(data_frame_copy_num)

#cambiamos los valores nulos por otro
data_frame_copy_num_nonan = imputer.fit_transform(data_frame_copy_num)
data_frame_copy = pd.DataFrame(data_frame_copy_num_nonan, columns=data_frame.columns)
print("Comprobamos si estan todas las columnas: " , data_frame_copy['Profession'].value_counts().sum())
data_frame_copy = data_frame_copy.drop("CustomerID", axis=1)
data_frame_copy
```

### 2.1 Matriz de correlación

```python
#construcción de la matriz de correlación
corr = data_frame_copy.corr()
fig, ax = plt.subplots(figsize=(8, 8))
ax.matshow(corr)
#plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show
```
